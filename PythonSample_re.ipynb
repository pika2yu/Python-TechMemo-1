```python
import re
```

```python
# search /match
#------------------
# match は string の先頭で 0 個以上の文字が正規表現 pattern とマッチすれと、対応する Match を
# 返す。文字列がパターンとマッチしない時は None を返す。
# 「先頭で」という点が search との違い。
#
# search は string 全体を走査して、正規表現 pattern がマッチを発生する最初の位置を探し、
# 対応する Match を返す。
#
def check_match(pattern, text):
    m = re.match(pattern, text)
    if m:
        print(f'一致 ({text})')
        #print( m.expand('data=\g<0>\n') )
        print( m.expand('data=\g<0>') )
    else:
        print(f'一致しない({text})')
    return m

def check_search(pattern, text):
    m = re.search(pattern, text)
    if m:
        print(f'一致 ({text})')
        #print( m.expand('data=\g<0>\n') )
        print( m.expand('data=\g<0>') )
    else:
        print(f'一致しない({text})')
    return m
```

```python
# pattern は ドライブレター(a:,b:,...)
pattern = r'[a-zA-Z]:'  # ドライブレター(a:,b:,...)

print('***match だと先頭にドライブレターがある時だけ一致')
check_match(pattern, 'a:okada')
check_match(pattern, 'drive a:okada')

print('\n')
print('*** search だと中間にドライブレターがある場合も一致')
check_search(pattern, 'a:okada')
check_search(pattern, 'drive a:okada')

#
# cf. 決まった文字列の時は in で済む
#
print('\n')
target = "OK"
if target in 'The result is OK.':
    print('OK is included')
else:
    print('OK is NOT included')
```

```python
# Match オブジェクトへのアクセス
# --------------------------
# (1). Match.gropu(index) で一致した結果のサブグループとして分割してアクセスできる。
# 正規表現中で()で囲むと、該当するそ部分が サブgroup になる。
# match.group() または index=0としてmatch.group(0) はマッチした文字列全体。
# (2). Match.expand(template) で整形を加えて結果にアクセスできる。
#
print('()で囲むと、それぞれの部分がサブgroupになる。')
pattern = '(\w+):(\w+):(\w+)'  # 英数字
match = check_search(pattern, 'It is 21:34:57')
    # 21 34 57  の３つのサブグループ
    # : は（）の外なのでサブグループでアクセスした時は含まれない
print(match.lastindex)
print(match.group())   # 全体の表示
print('サブグループでアクセスすると()の外の:は含まれない')
print(match.group(1))  # 最初の()
print(match.group(2))  # 2番目
print(match.group(3))  # 3番目
print('expand のtemplateのパターン')
print(match.expand('data=\g<0>'))      # \g<0> \g<1> 等でグループを表す
print(match.expand('data=\g<1>\n'))    # \n で改行
print(match.expand('data=\g<2>\n'))
print(match.expand('data=\g<3>\n'))

#---------
print('\n')
print('()で囲むと、グループとして定義されたそれぞれの部分の数分に該当するデータが無いと不一致。')
match = check_search(pattern, 'It is 21:34')
    # This と is の２つしかないので一致しない

#------
print('\n')
print('()で囲まないと group にならない。')
pattern = '\w+ \w+ \w+'
    # () で囲んでいないのでは This is a の全体で１グループ。サブグループはない。 
match = check_search(pattern, 'It is 21:34:57')
print(match.lastindex)  # サブグループがないので None
print(match.group(0))   # 全体の表示
print(match.group(1))   # サブグループがないので IndexError になる
```

```python
#--------------------
# findall
#-------------------- 
# findall はマッチする、全ての文字列を抽出する。戻りはStringのリスト。
# 全てをリストにする所が match / search と異なる。
#
# 抽出する範囲は正規表現の定義で調整できる。
#
# '\w+' のようにすると英数字の文字列のリスト
print('\w+ のようにすると英数字の文字列のリスト')
pattern = '\w+'
hitList=re.findall(pattern,'There are "apple", "orange" and "banana".')
for text in hitList:
    print(text)
    
# '"\w+"' のようにすると""に囲われた英数字のリスト
print('"\w+" のようにすると""に囲われた英数字のリスト')
pattern = '"\w+"'
hitList=re.findall(pattern,'There are "apple", "orange" and "banana".')
for text in hitList:
    print(text)

# さらに '"(\w+)"' のように () で囲って""の中身をグループ指定をすると""の中身だけのリスト
print('さらに "(\w+)" のように () で囲って""の中身をグループ指定をすると""の中身だけのリスト')
pattern = '"(\w+)"'
hitList=re.findall(pattern,'There are "apple", "orange" and "banana".')
for text in hitList:
    print(text)
```

```python
# () の使い方 : advanced
#---------------
# こんな使い方もある
# If multiple groups are present, return a list of tuples of strings
# matching the groups. Non-capturing groups do not affect the form of the result.
#
result=re.findall(r'(\w+)=(\d+)', 'set width=20 and height=10')

print(result)

for tuple1 in result:
    for text in tuple1:
        print(text)
```
