```python
#*****************************************
# 関数の宣言
#   def 関数名(パラメータ....):
#*****************************************
```

```python
# 関数は先頭に def をつける。最後は : をつける。
#   def 関数名(パラメータ....):
# { } などで括るのではなく、indent が終わるまでが一つの関数
def fnc_print_1(message):
    print(f'output from print 1 : {message}')
def fnc_print_2(message):
    return f'output from print 2 : {message}' # 戻りは普通にreturn
```

```python
fnc_print_1('Test Message')
ret = fnc_print_2('Test Message')
print(f'return value = {ret}')
```

```python
def fnc_split_parm_1(num1,  # １つ目と２つ目を改行。
        num2, \
                   # バックスラッシュで明示的な改行を示めしても良いが、同じ行にコメントはけけなくなる
        num3,      # バックスラッシュを使わなければコメントが書ける 
        num4):     # 
    sum = num1 + num2
    mul = num3 * num4
    print(f'{num1}+{num2}={sum}')
    print(f'{num3}x{num4}={mul}')
```

```python
fnc_split_parm_1(3,5,7,8)
```

```python
#*****************************************
# 変数のスコープ - 1/2
#*****************************************
#
# Python は変数の宣言が不要な代わりに、スコープに癖がある。
#
# 関数の中途でglobal variableを書き換えようとするとエラーになる。
#
```

```python
global_var1 = 'How are you ?'
def fnc_var():
    print(global_var1)

    # !!!! 要注意 !!!!
    # gloval valiable を書き換えようとするとエラーになる
    # 内部的には、ここで gloval valiable を local variableと認識して代入しようとする処理になりエラーになるらしい
    #global_var1 = 'Changed'  

    local_var = 'Good Morning!'
    print(local_var)
    
fnc_var()

# 当たり前だが、関数内の local value は参照できない。
#print(loval_var)
```

```python
#*****************************************
# 変数のスコープ - 2/2
#*****************************************
#
# 明示的にlocal/global variableである事をわからせてやると思ったように扱える。
# - gloval variable と同じ名前の変数を関数の頭で初期化すると local variable として扱われる。
# - 関数の中で gloval で宣言してやると global variable として扱える。
#
```

```python
global_var2 = 'How are you ?'
global_var3 = 'Good Morning!'
def fnc_var_explicit():
    #
    # 関数の先頭で代入してやると、名前が同じでも local value として扱われ、エラーにならない。
    #
    global_var2 = 'var2 local:Changed' 
    print(global_var2) # ここではlocalな値がプリントされる

    #
    # globalをつけて明示的に宣言するとglobal variableとして扱える。
    #
    global global_var3   # global として宣言
    print(global_var3)   # global の値が見えている
    global_var3 = 'Good Night'  # 書き換えはエラーにならず、glovalな値が書き換えられる
    
fnc_var_explicit()

print(f'var2 global: {global_var2}') # glovalな値は変わらない。

print(global_var3)                   # gloval な値が書き換えられている。
```
