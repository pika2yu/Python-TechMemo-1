```python
from glob import glob
import os
#
# File list の取得
#--------------------
#
# glob を使う
#
print('globで取得')
flist = glob ('*.*')
print(flist)
#
# glob で取れるのはファイル名。パスを付けたい場合はos.path.abspath()で変換
#
print()
print('os.path.abspath()で変換')
absfList = [os.path.abspath(f) for f in flist] # abspath を追加してリストを作成
print(absfList)
```

```python
#--------------------
# file I/O
#--------------------
#
# withブロックを使うことで、明示的にクローズしなくても、自動的に安全にファイルをクローズしてくれる。
#
with open('sample.txt', 'w', encoding='utf-8') as outfile:
    # ファイルのインスタンスを取得した後は、
    # read(), readline(), write() など、目的に応じて I/O を行う
    
    # 以下のように print に file= を指定して書き込む方法もある。
    print('1st line', file=outfile)
          
```
